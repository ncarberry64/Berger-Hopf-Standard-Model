# Phase 30 - Virtual Dressing Adoption Gate

This notebook evaluates whether the virtual-environment `1/2` dressing rule qualifies as an adoption candidate. It does not mark the rule as canonically adopted.

```python
import sys
from pathlib import Path

repo = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(repo / 'src'))

from bhsm_model import build_bhsm_model
from virtual_environment import adoption_report, bare_vs_dressed_prediction_ledger, pure_fiber_middle_up_rule

model = build_bhsm_model()
rule = pure_fiber_middle_up_rule()
report = adoption_report(model, rule)
report['rule']
```

## Adoption Criteria Table

```python
for criterion in report['criteria']:
    print(criterion['id'], criterion['passes'], criterion['statement'])
    print('  evidence:', criterion['evidence'])
```

## Bare vs Dressed Outputs

```python
for row in bare_vs_dressed_prediction_ledger(model, rule):
    print(row)
```

## Residual Comparison and Scope

```python
print('changed outputs:', report['changed_outputs'])
print('unrelated sectors changed:', report['unrelated_sectors_changed'])
for variant in report['model_variants']:
    print('\n', variant['name'])
    for row in variant['comparisons']:
        print(row)
```

## Audit Decision

If all criteria pass, the rule is marked `ADOPTION_CANDIDATE`, not `ADOPTED_CANONICAL_DRESSED`. Final adoption is reserved for a later explicit phase.

```python
print('all criteria pass:', report['all_criteria_pass'])
print('status:', report['rule_status_after_audit'])
print('adopted canonical dressed:', report['adopted_canonical_dressed'])
```
