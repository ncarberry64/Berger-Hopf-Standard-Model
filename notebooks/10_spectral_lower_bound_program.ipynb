# Gate 32B - Spectral Lower-Bound Program

This notebook audits sufficient lower-bound inequalities for the Level 2 finite-basis twisted Dirac proxy. It does not prove the no-extra-light-state theorem and does not compute the full analytic `H_T` spectrum.

## No-Extra-Light-State Inequality

The target condition is

```text
H_T restricted to H_perp >= (4 pi^2 v)^2
```

In the dimensionless heat-lift audit the sufficient lower-bound condition is

```text
d_lower + mu_H * (1 - exp(-d_lower / Lambda^2)) + V_min >= mu_H
```

with the natural cutoff `Lambda^2 = 1/(4 pi)`.

## Limitations

- `DIRAC_PROXY_LEVEL_2` is representation-aware and matrix-based, but still finite-basis.
- Gershgorin, min-max, and Weyl checks are sufficient finite-matrix lower-bound estimates.
- The complete Hilbert-space complement `H_perp` is not controlled here.
- The no-extra-light-state theorem remains open until the full analytic `H_T` spectrum or a rigorous complete-space lower bound is obtained.
