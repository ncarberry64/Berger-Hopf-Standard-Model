## Gate 25B — Boundary-Operator Mode Selection

Boundary operators:

- `q = k - 2j`
- `Omega_l = 2j - q`, target `3`, with `q = 1 mod 2`
- `Omega_u = q - 2j`, target `6`, with `q = 0 mod 2` and `q >= 6`
- `Omega_d = q + 4j`, target `12`, with `q = 0 mod 4`

The scan uses the nonnegative Hopf-charge Berger domain `0 <= j <= floor(k/2)`. The heavy mode `(0,0)` is included separately.

```python
admissible_sublattices = {
    sector: [
        {'mode': mode, 'lambda': berger_lambda(*mode)}
        for mode in admissible_modes(sector, k_max=20)[:5]
    ]
    for sector in ['lepton', 'up', 'down']
}
admissible_sublattices
```

```python
mode_ledger(k_max=12)
```

```python
robustness_table = []
for k_max in [10, 12, 20, 40]:
    report = selection_report(k_max)
    robustness_table.append({
        'k_max': k_max,
        'recovered_all': report['recovered_all'],
        'lepton': report['sectors']['lepton']['selected'],
        'up': report['sectors']['up']['selected'],
        'down': report['sectors']['down']['selected'],
        'lower_action_competitors': {
            sector: report['sectors'][sector]['lower_action_competitors']
            for sector in ['lepton', 'up', 'down']
        },
    })
robustness_table
```

Gate 25B is an operational audit only. It shows that the supplied boundary operators recover the charged-sector mode ledger without observed mass inputs. It is not yet a derivation of `Omega_f` from the full twisted Dirac or bundle action.
