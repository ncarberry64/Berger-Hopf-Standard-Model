# Phase 25 - Quark Mass Scheme and Running Audit

This notebook audits the comparison scheme for BHSM charged-quark mass ratios. It does not tune BHSM parameters, change canonical geometry, fetch external data, or implement QCD running.

```python
import sys
from pathlib import Path

repo = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(repo / 'src'))

from bhsm_model import build_bhsm_model, compute_yukawa_ratios
from mass_scheme import default_mass_references, mass_scheme_report, compare_bhsm_ratios_to_schemes

model = build_bhsm_model()
print(model.geometry_config)
compute_yukawa_ratios(model)
```

## Current Mass Reference Table

```python
refs = default_mass_references()
for scheme, rows in refs.items():
    print('\n', scheme)
    for particle, ref in rows.items():
        print(particle, ref.value, ref.unit, ref.scheme, ref.scale)
```

## Scheme Consistency Flags

```python
report = mass_scheme_report()
for scheme, rows in report['ratio_references'].items():
    print('\n', scheme)
    for key, row in rows.items():
        print(key, row['ratio'], row['scheme_consistent'], row['scheme'], row['scale'])
```

## BHSM Ratios vs MIXED_DEFAULT

```python
for row in compare_bhsm_ratios_to_schemes(model, 'MIXED_DEFAULT'):
    print(row['id'], 'pred=', row['predicted'], 'ref=', row['reference'], 'rel=', row['relative_error'], 'scheme_sensitive=', row['scheme_sensitive'])
```
