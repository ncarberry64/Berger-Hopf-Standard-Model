# Phase 26 - Common-Scale Quark Mass Running Scaffold

This notebook applies an approximate one-loop-inspired running scaffold to quark mass references. It does not tune BHSM parameters and does not claim precision QCD matching.

```python
import sys
from pathlib import Path

repo = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(repo / 'src'))

from bhsm_model import build_bhsm_model, compute_yukawa_ratios
from mass_scheme import mass_scheme_report
from quark_running import MZ, MT_LIKE_SCALE, TEN_GEV, alpha_s_one_loop, compare_bhsm_to_common_scale, quark_running_report

model = build_bhsm_model()
print(model.geometry_config)
compute_yukawa_ratios(model)
```

## Current Mixed References

```python
report = mass_scheme_report()
for key, row in report['ratio_references']['MIXED_DEFAULT'].items():
    if 'quarks' in key:
        print(key, row['ratio'], row['scheme'], row['scale'], row['scheme_consistent'])
```

## Approximate Running Formula

`m(mu_2) = m(mu_1) [alpha_s(mu_2)/alpha_s(mu_1)]^p`, with fixed `nf=5` and `p=12/23`. This is an `APPROXIMATE_RUNNING_SCAFFOLD`, not precision QCD.

```python
for scale in (TEN_GEV, MZ, MT_LIKE_SCALE):
    print(scale, alpha_s_one_loop(scale))
```

## Common-Scale Ratio Tables

```python
running = quark_running_report(model)
for scale, rows in running['comparisons'].items():
    print('\nTarget', scale)
    for row in rows:
        print(row['id'], 'ref=', row['reference'], 'bhsm=', row['predicted'], 'rel=', row['relative_error'])
```

## Target-Scale Controls

```python
for target in (MZ, MT_LIKE_SCALE, TEN_GEV):
    rows = compare_bhsm_to_common_scale(model, target)
    print('\n', target)
    for row in rows:
        print(row['id'], row['relative_error'], row['status'])
```

## Limitation

Full QCD running, threshold matching, higher-loop evolution, and uncertainty propagation remain open. This scaffold only separates common-scale comparison infrastructure from the canonical BHSM predictions.
