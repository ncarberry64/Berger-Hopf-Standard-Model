# Phase 23 - BHSM Canonical Geometry Audit

This notebook audits the available Berger anisotropy configurations without tuning empirical residuals. The adoption rule is theory-side: if the BHSM assumptions include `epsilon_alpha = alpha^{-1}/(12*pi^2) - 1`, the canonical model uses the alpha-anchored Berger geometry.

```python
import sys
from pathlib import Path

repo = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(repo / 'src'))

from bhsm_config import available_geometry_configs, canonical_geometry_config, compare_geometry_configs
from bhsm_model import build_bhsm_model

configs = available_geometry_configs()
canonical = canonical_geometry_config()
canonical
```

## Geometry Configurations

`ROUND` remains a baseline control, `LEGACY_LOW_A` remains sensitivity-only, and `ALPHA_ANCHORED` is selected only when the alpha/Hopf residual rule is admitted.

```python
for config in configs:
    print(config)
```

## Full-Ledger Comparison

The comparison recomputes charged mass ratios, CKM angles, PMNS effective-extension rows, gauge couplings, Higgs/electroweak outputs, and residual summaries for each geometry.

```python
rows = compare_geometry_configs(build_bhsm_model, configs)
for row in rows:
    g = row['geometry']
    up = row['up_sector']
    worst = row['worst_residuals'][0]
    print(g['name'], 'a=', g['a'])
    print('  c/t =', up['c_over_t'])
    print('  u/t =', up['u_over_t'])
    print('  sin theta_13 =', up['sin_theta_13'])
    print('  worst residual =', worst)
```

## Up-Sector and CKM Vub Comparison

The alpha-anchored geometry improves the localized up-sector and `sin(theta_13)` diagnostic, but that improvement is not the adoption criterion. The adoption criterion is the internal geometry rule tied to `epsilon_alpha`.

```python
comparison = [
    {
        'geometry': row['geometry']['name'],
        'a': row['geometry']['a'],
        'c_over_t': row['up_sector']['c_over_t'],
        'u_over_t': row['up_sector']['u_over_t'],
        'sin_theta_13': row['up_sector']['sin_theta_13'],
        'worst_prediction_id': row['worst_residuals'][0]['prediction_id'],
        'worst_relative_error': row['worst_residuals'][0]['relative_error'],
    }
    for row in rows
]
comparison
```
